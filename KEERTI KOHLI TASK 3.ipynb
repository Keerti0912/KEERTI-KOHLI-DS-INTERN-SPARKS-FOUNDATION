```python
st_profit=SampleSuperstore_data.groupby(["State"])["Profit"].sum().nlargest(20)
```

```python
st_profit
```

```python
plt.figure(figsize=(10,10))
st_profit.plot.bar()
```

```python
sns.lineplot(data=SampleSuperstore_data, x="Discount", y= "Profit")
```

```python
# profit and discount are inversely proportional
# as discount rises profit falls
```

```python
SampleSuperstore_data.plot(kind="scatter",x="Sales",y="Profit", c="Discount", colormap="Set1",figsize=(10,10))
```

```python
data1= SampleSuperstore_data.groupby("State")[["Sales","Profit"]].sum().sort_values(by="Sales", ascending=False)
data1[:].plot.bar(color = ["Blue","Yellow"], figsize=(10,15))
plt.title("Profit-Loss and Sales in States")
plt.show()
```

```python
# CALIFORNIA and NEWYORK generates the highest profit
```

```python
data1= SampleSuperstore_data.groupby("Region")[["Sales","Profit"]].sum().sort_values(by="Sales", ascending=False)
data1[:].plot.bar(color = ["Red","Green"], figsize=(10,7))
plt.title("Profit-Loss and Sales across Region")
plt.show()
```

```python
#Conclusion: 
#1. Central region generates the lowest profit.
#2. Western region yields the highest profit.
#3. Ohio, Texas are having loss.
#4. We have to work more in the weak areas in Ohio , Texas etc to make more profit.
```
